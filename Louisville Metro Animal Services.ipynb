## The Barchart Shows What breed was the most taken in by LMAS 2005-2017

```python
#Counts and makes a very basic bar graph to showing what breeds where most taken in by 
#Louisville metro animal services from 2005-2017

res = Q("""
SELECT
   (PrimaryBreed) as Breed, 
   count(*) as Count
   from animals
   group by PrimaryBreed
   ORDER BY Count desc limit 20

""")
res.plot.bar(x = "Breed")
```
